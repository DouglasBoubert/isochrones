```python
from scipy.stats import uniform

from isochrones import get_ichrone
from isochrones.priors import ChabrierPrior, FehPrior
from isochrones.populations import StarFormationHistory, StarPopulation

mist = get_ichrone('mist')
sfh = StarFormationHistory(uniform(0, 10))  # Constant SFR for 10 Gyr 
imf = ChabrierPrior(bounds=(0.8, 10))  # bounds on solar masses
feh = FehPrior(halo_fraction=0.2)
distance = 1000  # pc  (or this can be an isochrones.prior.Prior object, or anything with a "sample" method, really)
AV = 0.  # Same rules as distance

pop = StarPopulation(mist, sfh=sfh, imf=imf, feh=feh, distance=distance, AV=AV)

df = pop.generate(1e4)
```

```python
import holoviews as hv
hv.extension('bokeh')
import hvplot.pandas

df['G-J'] = df.G_mag - df.J_mag

options = dict(invert_yaxis=True)
hv.Layout([df.hvplot.scatter('G-J', 'G_mag', size=4, alpha=0.5, color='feh', cmap='viridis').options(**options), 
           df.hvplot.scatter('G-J', 'G_mag', size=4, alpha=0.5, color='age', cmap='viridis').options(**options)]).cols(1)
```
